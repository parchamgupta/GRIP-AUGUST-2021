##Importing the required libraries

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import cv2
from collections import Counter
import os
%matplotlib inline
```

##Defining function to return the hexadecimal code for the specific colour

```python
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))
```

##Defining function to read the image and convert it from BGR to RGB as CV2 reads images in BGR

```python
def get_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

##Defining function to perform kmeans clustering on the images to identify different colour clusters. 
##K is taken as 10 for optimal results

```python
def get_colors(image, number_of_colors, show_chart):
    modified_image = cv2.resize(image, (960, 540), interpolation = cv2.INTER_AREA)
    modified_image = modified_image.reshape(modified_image.shape[0]*modified_image.shape[1], 3)
    clf = KMeans(n_clusters = number_of_colors)
    labels = clf.fit_predict(modified_image)
    counts = Counter(labels)

    center_colors = clf.cluster_centers_
    # We get ordered colors by iterating through the keys
    ordered_colors = [center_colors[i] for i in counts.keys()]
    hex_colors = [RGB2HEX(ordered_colors[i]) for i in counts.keys()]
    rgb_colors = [ordered_colors[i] for i in counts.keys()]

    if (show_chart):
        plt.figure(figsize = (10, 8))
        plt.pie(counts.values(), labels = hex_colors, colors = hex_colors)

    return rgb_colors
```

##Running the function on first image

```python
plt.figure(figsize=(10, 8))
image = get_image('../content/sample.jpg')
print("The type of this input is {}".format(type(image)))
print("Shape: {}".format(image.shape))
plt.imshow(image)
```

##Piechart containing proportions of different colours in the first image

```python
get_colors(image,10,True)
```
